```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow # Import the cv2_imshow function from google.colab.patches
import torch
import os
```

# Image Processing Technique
- **Grayscale Conversion:** Converts an image from RGB to grayscale, reducing complexity.
- **Resizing:** Changes the dimensions of an image while maintaining aspect ratio.
- **Normalization:** Scales pixel values (e.g., between 0 and 1 or -1 and 1) for uniformity.
- **Noise Reduction:** Filters (e.g., Gaussian blur) remove unwanted noise from images.

# Load the sample image

```python
# reading the image
cv_img = cv.imread("/content/sami.jpg")
cv_img
```

```python
cv2_imshow(cv_img)
```

```python
# Resize the img
resize_img = cv.resize(cv_img, (224,224))
cv2_imshow(resize_img)
```

```python
# Convert the img to gray scale
gray_img = cv.cvtColor(resize_img, cv.COLOR_BGR2GRAY)
gray_img
```

```python
cv2_imshow(gray_img)
```

```python
# Normalize the image
norm_img = gray_img/255
norm_img
```

```python
cv2_imshow(norm_img)
```

```python
# Noice Reduction
cv.GaussianBlur(resize_img , (5,5),0)
```

# Image Enhancement

```python
# Histogram Equalization
cv.equalizeHist(gray_img)
```

```python
# Sharpenig
cv.filter2D(gray_img, -1, np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]]))
```

```python
# smoothing
cv.blur(gray_img, (5,5))
```

```python
# Gamma Correction
img =cv.pow(gray_img.astype(np.float32), 0.5)
cv2_imshow(img)
```

# Feature Extraction

```python
# Edge Detection
cv.Canny(gray_img, 100, 200)
```

```python
# Contours
contours, hierarchy = cv.findContours(gray_img, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
```

```python
cv.drawContours(gray_img, contours, -1, (0, 255, 0), 3)
```

```python
# Corner Detection
corners = cv.goodFeaturesToTrack(gray_img, 100, 0.01, 10)
corners
```

```python
cv.circle(gray_img, (int(corners[0][0][0]), int(corners[0][0][1])), 5, (0, 0, 255), -1)
```

```python
# Blob Detection
params = cv.SimpleBlobDetector_Params()
params.filterByArea = True
params.minArea = 100
detector = cv.SimpleBlobDetector_create(params)
keypoints = detector.detect(gray_img)
```

```python
cv.drawKeypoints(gray_img, keypoints, gray_img, (0, 0, 255), cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```
